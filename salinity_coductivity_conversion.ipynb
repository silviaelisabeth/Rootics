```python
# parameter settings
temp_degC = 13  # temperature in degC
atm = 1.01325   # atmospheric pressure in bar | convert bar in decibar: 1bar == 10dbar
cnd = 1         # conductivity ratio (for salinity 35, temp_degC=15, atm -> cnd = 1)
calc = 0        # which calculation direction to choose: calc=0 --> conductivity ratio to salinity; calc=1 --> salinity to conductivity ratio

if calc == 0:
    print('conductivity ratio', cnd,'S*m-1 to salinity @', temp_degC, 'degC and', 10/1*atm, 'dbar:')
    unit = 'PSU'
else:
    print('salinity', cnd,'PSU to conductivity ratio @', temp_degC, 'degC and', 10/1*atm, 'dbar:')
    unit = 'S*m-1'
print('result', round(SalCon_Converter(temp_degC=temp_degC, p_dbar=10/1*atm, M=calc, cnd=1), 4), unit)
```
